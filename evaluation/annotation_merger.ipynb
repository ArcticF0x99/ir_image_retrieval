```python
import pandas as pd
from glob import glob
import os
from typing import List, Set
```

```python
def add_merged_annotation(out_path : str, filename : str, annotation : set):
    with open(out_path, "a") as f:
        f.write(f"{filename}, {repr(annotation)}\n")

def count_occurances(labels: List[Set[str]], label : str):
    count = 0
    for l in labels:
        if label in l:
            count += 1
    return count

def check_annotation(annotations_glob : str, out_path : str):
    dfs = [pd.read_json(path, lines=True) for path in glob(annotations_glob)]
    if len(dfs) != 3:
        raise Exception('Only built for 3 annotators')

    for index in range(dfs[0].shape[0]):
        filename = dfs[0].iloc[index]['filename']
        labels = [set(df.iloc[index]['label']) for df in dfs]
        has_consens = len(set(repr(label) for label in labels)) == 1
        if has_consens:
            add_merged_annotation(out_path, filename, list((labels[0] - {'On Topic'}))[0])
            continue

        off_topic_count = count_occurances(labels, 'Off Topic')
        neutral_count = count_occurances(labels, 'Neutral')
        pro_count = count_occurances(labels, 'Pro')

        if off_topic_count >= 2:
            add_merged_annotation(out_path, filename, 'Off Topic')
        elif off_topic_count == 1:
            # 2 On Topic
            if neutral_count == 0:
                if pro_count == 2:
                    add_merged_annotation(out_path, filename, 'Pro')
                elif pro_count == 1:
                    add_merged_annotation(out_path, filename, 'Neutral')
                elif pro_count == 0:
                    add_merged_annotation(out_path, filename, 'Con')
            elif neutral_count >= 1:
                add_merged_annotation(out_path, filename, 'Neutral')
        elif off_topic_count == 0:
            if neutral_count == 0:
                if pro_count >= 2:
                    add_merged_annotation(out_path, filename, 'Pro')
                elif pro_count == 1:
                    add_merged_annotation(out_path, filename, 'Con')
            elif neutral_count == 1:
                if pro_count == 2:
                    add_merged_annotation(out_path, filename, 'Pro')
                elif pro_count == 1:
                    add_merged_annotation(out_path, filename, 'Neutral')
                elif pro_count == 0:
                    add_merged_annotation(out_path, filename, 'Con')
            elif neutral_count == 2:
                add_merged_annotation(out_path, filename, 'Neutral')


out_path = 'merged_annotation.csv'
try:
    os.remove(out_path)
except OSError:
    pass


check_annotation('annotations/*.jsonl', out_path)
```
